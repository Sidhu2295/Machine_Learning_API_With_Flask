```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("london_housing.csv", encoding='unicode_escape')
df.head()
```

```python
prices = list(df.Price)
```

```python
import re

correct_prices = []
for price in prices:
    test = re.sub("[^\d\.]", "", price)
    correct_prices.append(int(test))
```

```python
df['Price'] = correct_prices
df.head()
```

```python
df.info()
```

```python
df.Location.value_counts()
```

```python
x = df.Location
y = df.Price

plt.figure(figsize=(8,6))
ax = sns.scatterplot(data=df, x=x, y=y)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha='right')
plt.tight_layout()
plt.show()
```

```python
df2 = df.drop(['Address','Listing_Type'], axis=1)
df2['Location'] = df2['Location'].replace(['Abbey_Wood'], 1)
df2.head()
```

```python
x = df2['Bedroom']
y = df2.Price

sns.scatterplot(data=df2, x=x, y=y)
```

```python
df2['Bedroom'].astype(float)
df2['Bathroom'].astype(float)
df2['Living_Room'].astype(float)
```

```python
x = df2[['Bedroom']]
y = df2.Price

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver='lbfgs', max_iter=1000)

lr.fit(X_train, y_train)
```

```python
y_predicted = lr.predict(X_test)
```

```python
y_predicted
```

```python
y_test
```

```python
print('Accuracy: {:.2f}'.format(lr.score(X_test, y_test)))
```
